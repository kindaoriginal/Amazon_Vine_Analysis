```python
import pandas as pd
from path import Path
```

```python
path = Path('./data/vine_table.csv')
df = pd.read_csv(path)
# df.columns = ['review_id','star_rating','helpful_votes','total_votes','vine', 'verified_purchase']
```

```python
# new df for total_votes greater than 20
total_votes_df = df.loc[df['total_votes'] >= 20]
total_votes_df.sort_values(by='total_votes', ascending=True)
```

```python
# new df for helpful_votes percentage gerater than 50%
helpful_votes = (total_votes_df['helpful_votes']/total_votes_df['total_votes'])
helpful_votes_df = total_votes_df.loc[(helpful_votes >= 0.50)]
helpful_votes_df.head()
```

```python
paid_df = helpful_votes_df.loc[df['vine'] == 'Y']
unpaid_df = helpful_votes_df.loc[df['vine'] == 'N']
```
